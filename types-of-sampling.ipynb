```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
```

```python
a_g_dataset=pd.read_csv("/kaggle/input/sampling/age_gender.csv")
a_g_dataset.info()
```

```python
#dataset filter
age_data=a_g_dataset.dropna(subset=['age'])
age_data=age_data[age_data["age"]>0]
sample_dataset_for_sampling,_=train_test_split(age_data,test_size=0.8,stratify=age_data['gender'],random_state=50)
print(len(sample_dataset_for_sampling))
```

```python
sns.set(style='whitegrid')#Canvas color style
plt.figure(figsize=(8,4)) #canvas area
sns.histplot(data=sample_dataset_for_sampling, 
             x='age',
             hue='gender',
             bins=100,
             kde=True,
             palette='Set2',
             stat='density',
             common_norm=False)
plt.title('Stratify Data Sampling')
plt.xlabel('Age Datasets')
plt.ylabel('Gender Datasets')
plt.legend(title='Gender',
          labels=['Female(0)','Male(1)'])
plt.tight_layout()
plt.show()
```

# Systemic Sampling
* Formula, K=N/n
* Where, N=Total Population; n= Desired sample size

```python
credit_data=pd.read_csv("/kaggle/input/sampling-datasets/credit_data.csv")
credit_data.head(2)
#credit_data.info()
```

```python
#dataset filter
credit_data=credit_data.dropna(subset='age')
nth_interval_value=50
starting_point=np.random.randint(0,nth_interval_value)
print(starting_point)
systemic_sampling=credit_data.iloc[starting_point::nth_interval_value]
print(systemic_sampling)

sns.set(style='whitegrid')
plt.figure(figsize=(8,4))
sns.histplot(systemic_sampling['age'],
            bins=30,
            kde=True,
            palette='Set2')
plt.title('Systemic Data Sampling')
plt.xlabel('Age datasets')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
